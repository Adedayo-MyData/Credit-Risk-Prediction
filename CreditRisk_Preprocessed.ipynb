# CREDIT RISK DATA PREPROCESSING

## Import Relevant Libraries

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

## Loading the Raw Data

```python
raw_csv_data = pd.read_csv("credit_risk_raw.csv")
data = raw_csv_data.copy()
```

## Data Preprocessing

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.dropna(axis=0)
```

```python
data.info()
```

```python
data_without_ID=data.drop(['loan_id'],axis=1)
```

```python
data_without_ID.head()
```

Multicollinearity Check

```python
data_without_ID.columns.values
```

```python
variables = data_without_ID[['annual_income', 'monthly_debt', 'loan_amount',
       'total_credit_limit', 'credit_used', 'total_delinquencies_12m',
       'open_credit_accounts', 'credit_history_years',
       'oldest_account_age', 'late_payments_90d',
       'hard_inquiries_last_6m']]

vif = pd.DataFrame()
vif['Feature'] = variables.columns
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
```

```python
vif
```

## Features Engineering

```python
data_without_ID["debt_to_income"] = data_without_ID["monthly_debt"] / (data_without_ID["annual_income"] / 12)
```

```python
data_without_ID["credit_utilization"] = data_without_ID["credit_used"] / data_without_ID["total_credit_limit"]
```

```python
data_without_ID["loan_to_income"] = data_without_ID["loan_amount"] / data_without_ID["annual_income"]
```

```python
data_without_ID["delinquencies_last_12m"] = data_without_ID["total_delinquencies_12m"]
```

```python
data_without_ID["delinquency_ratio"] = data_without_ID["delinquencies_last_12m"] / data_without_ID["open_credit_accounts"]
```

```python
data_without_ID["utilization_to_income"] = data_without_ID["credit_utilization"] / data_without_ID["annual_income"]
```

```python
data_without_ID.columns
```

```python
data_without_ID=data_without_ID.drop(['annual_income', 'monthly_debt', 'loan_amount', 'total_credit_limit',
       'credit_used', 'total_delinquencies_12m'],axis=1)
data_without_ID.columns
```

```python
columns_reordered=['late_payments_90d', 'credit_history_years', 'oldest_account_age',
       'open_credit_accounts', 'hard_inquiries_last_6m',
       'debt_to_income', 'credit_utilization', 'loan_to_income',
       'delinquencies_last_12m', 'delinquency_ratio', 'utilization_to_income', 'default']
```

```python
data_without_ID=data_without_ID[columns_reordered]
```

```python
data_without_ID.head()
```
